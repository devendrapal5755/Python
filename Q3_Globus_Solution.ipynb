```
A= input("Please enter value of A : ")
B= input("Please enter value of B : ")
C= input("Please enter value of C : ")
print(f"{A} + {B} = {C}")
```

```
if (A.isdigit() & B.isdigit() == True ):
  print (f"{A} + {B} = {int(A)+int(B)}")
elif (A.isdigit() & C.isdigit() == True):
  print (f"{A} + {int(C)-int(A)} = {C}")
elif (B.isdigit() & C.isdigit() == True):
  print (f"{int(C)-int(A)} + {B} = {C}")
```
