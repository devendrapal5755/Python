```
num = 0
visited = [0 for i in range(1001)]
adj = [[] for i in range(1001)]

def low(lt,a):
    if lt[-1] <= a:
        return len(lt)-1
    if lt[0] >= a:
        return 0
    for i,j in enumerate(lt):
        if j > a:
            return i-1

def cycles(i):
    global mini, l, visited, s, k, adj
    mini = min(mini,l[i])
    visited[i] = 1
    s += l[i]
    k += 1
    for ad in adj[i]:
        if visited[ad]:
            return
        else:
            cycles(ad)

while True:
    n = int(input())
    if n == 0:
        break
    
    num += 1
    
    l = list(map(int,input().strip().split()))
    lt = list(l)
    for i in range(1001):
        visited[i] = 0
        adj[i] = []
    lt.sort()

    for i in range(n):
        adj[i].append(low(lt,l[i]))
    s = 0
    for i in range(n):
        if not visited[i]:
            mini = 10*30
            k = 0
            cycles(i)
            s += (k-2)*mini
            if 2*(lt[0]+mini) < (mini-lt[0])*(k-1):
                s -= (mini-lt[0])*(k-1) - 2*(lt[0]+mini)
            
    print("Case",str(num)+":",s,end='\n\n')
```
